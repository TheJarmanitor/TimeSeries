```python
import pandas as pd
import numpy as np
import datetime
```

```python
data_path="D:/Documents/Jupyter Notebooks/TimeSeries/Proyect/Dataset/"
df=pd.read_csv(data_path+"DelayedFlights.csv").drop("Unnamed: 0", axis="columns")
df.head()
```

```python
df=df[["Year", "Month", "DayofMonth", "ArrTime", "CRSArrTime"]].dropna()
df["ArrTime"]=df["ArrTime"].astype(int)
df.head()
```

```python
df["TimeDifference"]=df["ArrTime"]-df["CRSArrTime"]
df.head()
```

```python
z=[]
for i in df["ArrTime"]:
    if len(str(i))==1:
        z.append("000"+str(i))
    elif len(str(i))==2:
        z.append("00"+str(i))
    else:
        z.append(str(i))
```

```python
df["hour"]=[int(i[:-2]) for i in z]
df["minute"]=[int(i[-2:]) for i in z]
```

```python
df.head()
```

```python
df=df.rename(columns={'Year':'year', 'Month':'month', 'DayofMonth':'day'})
df["Datetime"]=pd.to_datetime(df[["year", "month", "day", "hour", "minute"]])
df=df[["Datetime", "TimeDifference"]].set_index("Datetime")
```
